```python
# Wavelength data for framing a dataframe, instead of pixspec
i = 0
count = 0
wavelength_temp = 0

# Excluded wavelength ranges as described from the paper
big_range = range(10210, 26483)
# excluded_range_1 = range(16447, 17303)
# excluded_range_2 = range(19741, 20600)

# Creating a dataframe for wavelength for Ratioed spectra
df_Wavelength = pd.DataFrame()

# Loop to create the dataframe
while i < len(df_WL):
  wavelength_temp =  int(df_WL.iloc[i][0] * 10000)
  # print(wavelength_temp)
  # Check if the value is not in either of the excluded wavelength ranges
  if wavelength_temp in big_range:
    df_Wavelength.loc[:, f'{count}'] = [wavelength_temp / 10000]
    count +=1
  i += 1  # Increment the counter if the value is in the considered range but not in the excluded ranges

# Results for verification
print(i)
print(count)
print(wavelength_temp)

df_Wavelength
```
